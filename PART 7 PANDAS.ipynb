```python
x=0
y=[1, 1]
for c in range (2, 10):
    d=int(y[x])+int(y[x+1])
    y.append(d)
    x+=1

c=[]
for a in range (1, 25):
    sayac=0
    if a<=2:
        c.append(a)
    elif a%2!=0:
        for b in range(2, a//2):
            if (a%b)==0:
                sayac+=1
        if sayac==0:
            c.append(a)
indeks=["A", "B", "C", "Ç", "D", "E", "F", "G", "Ğ", "H"]
ind_a=pd.DataFrame(index=indeks)
ind_a.index.name="İND"
ind_a["asal_sayı"]=c
ind_a["fibonacci"]=y
ind_a
```
