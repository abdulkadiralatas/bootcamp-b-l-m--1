```python
plt.figure(figsize=(8,5))
n1 = 50

ortalama1 = []
for i in range(1, n1):
    a = np.random.normal(12, 3, 1000)
    ortalama1.append(np.average(a))
plt.hist(ortalama1, alpha=0.5, bins=50)
plt.gca().set_title('Popülasyon 1')
plt.figure(figsize=(8,5))

plt.show()


plt.figure(figsize=(8,5))
n2= 1000

ortalama2 = []
for i in range(1, n2):
    b = np.random.normal(100, 5, 1000)
    ortalama2.append(np.average(b))
plt.gca().set_title('Popülasyon 2')
plt.hist(ortalama2, alpha=0.5, bins=50)
plt.show()

# 50 Ortalama aldığımız Popülasyon1 normal dağılama çok yakın değil.
# 1000 Ortalama aldığımız Popülasyon2 grafiğini incelediğimizde normal dağılama çok yakın olmadığı anlaşılmaktadır.
```
